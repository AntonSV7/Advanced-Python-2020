```python
from os.path import isdir, join as join_path
from functools import partial
from warnings import filterwarnings
from os import mkdir
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
filterwarnings('ignore')
from scipy.stats import chisquare
DATA_DIR = 'Plazma'
to_data_dir = partial(join_path, DATA_DIR)
if not isdir(DATA_DIR):
    mkdir(DATA_DIR)
```

```python
plt.figure(figsize=(15,12))
x = data_1['I']
y = data_1['U']/9000*10**6
left_x = []
left_y = []
right_x = []
right_y = []
for i in range(3):
    left_x.append(x[i])
    left_y.append(y[i])
for j in range(len(y)-10,len(y)):
    right_x.append(x[j])
    right_y.append(y[j])
poly1 = np.polyfit(left_x,left_y,1)# 
poly2 = np.polyfit(right_x,right_y,1)
plt.plot(left_x, np.polyval(poly1,left_x),)
plt.plot(right_x, np.polyval(poly2,right_x),)
plt.scatter(x, y)
plt.xlabel('$U, V$',fontsize = 20)
plt.ylabel('$I,\mu A$',fontsize = 20)
plt.title('$250V, 48 \mu bar$',fontsize = 20)
plt.legend(fontsize = 20)
plt.grid()
plt.savefig(to_data_dir('1.png'))
print((poly1[0]*e/k+poly2[0]*e/k)/2)
print(poly1,poly2)
```
